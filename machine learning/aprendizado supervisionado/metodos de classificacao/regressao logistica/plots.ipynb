```python
# Importando bibliotecas

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.formula.api as smf
import statsmodels.api as sm
```

```python
# Gerando dados

X = np.linspace(-5, 5, 50)
y = (X>0).astype(float)
Y_log = 1 / (1 + np.exp(-X))

data = pd.DataFrame(X, y).reset_index().rename(columns = {'index': 'y', 0: 'x'})
```

```python
# Modelo Linear

linreg = smf.ols('y ~ x', data = data).fit()
b, a = linreg.params
Y = b + a*X
```

```python
# Plotando

fig, ax = plt.subplots(figsize = (7,5))

ax.scatter(X,y)
ax.plot(X, Y, c = 'red')

ax.set_title('Regressão linear com variável resposta binária', fontsize = 13, loc = 'left')
ax.set_xlabel('x', fontsize = 13)
ax.set_ylabel('y', fontsize = 13)

plt.show()
```

```python
fig.savefig('imagens/bin-lin.png')
```

```python
# Plotando

fig1, ax1 = plt.subplots(figsize = (8,5))

label = r'$sig(x) = \frac{1}{1 + e^{-x}}$'

ax1.scatter(X,y)
ax1.plot(X,Y_log, color = 'r', label = label)

#ax2.axhline(y = 1, lw = 1, ls = '--')
#ax2.axhline(y = 0, lw = 1, ls = '--')

ax1.grid(alpha = .3, ls = '--')

#ax2.margins(0)

#fig1.suptitle(t = 'Modelo Linear Generalizado', fontsize = 15)
ax1.set_title('Modelo Linear Generalizado', fontsize = 13, loc = 'left')
ax1.set_ylabel('sig(x)', fontsize = 13, loc = 'top', rotation = 0)
ax1.set_xlabel('x', fontsize = 13)
fig1.subplots_adjust()

#ax1.legend()

plt.show()
```

```python
fig1.savefig('imagens/bin-logit.png')
```

```python
# Plotando

fig2, ax2 = plt.subplots(figsize = (8,5))

label = r'$sig(z) = \frac{1}{1 + e^{-z}}$'

ax2.plot(X,Y_log, color = 'r', label = label)

#ax2.axhline(y = 1, lw = 1, ls = '--')
#ax2.axhline(y = 0, lw = 1, ls = '--')

ax2.spines['left'].set_position('center')
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)


ax2.grid(alpha = .3, ls = '--')

ax2.margins(0)

fig2.suptitle(t = 'Função Logística', x = .15, fontsize = 15)
ax2.set_ylabel('sig(z)', fontsize = 13, loc = 'top', labelpad = -20, rotation = 0)
ax2.set_xlabel('z', fontsize = 13)
fig2.subplots_adjust()

ax2.legend()

plt.show()
```

```python
fig2.savefig('imagens/sigmoide.png')
```

```python
# Plotando com threshold

fig3, ax3 = plt.subplots(figsize = (7,5))

ax3.scatter(X, y)
ax3.plot(X, Y_log, c = 'red')

ax3.set_title('Regressão linear generalizada com limite de decisão', fontsize = 13, loc = 'left')
ax3.set_xlabel('x', fontsize = 13)
ax3.set_ylabel('y', fontsize = 13)
#ax3.set_ylim(top = 1.2)

ax3.axhline(y = .5, ls = '--', lw = 1, c = 'black')
ax3.fill_between(x = X, y1=.5, y2 = -.1, color = 'red', alpha = .15)
ax3.fill_between(x = X, y1 = 1.1, y2 = .5, color = 'b', alpha = .15)


ax3.margins(0)

plt.show()
```
