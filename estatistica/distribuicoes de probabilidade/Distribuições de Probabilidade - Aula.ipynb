### <font color = #21618C> O que é uma probabilidade ?

Quando falamos sobre probabilidade em estatística, estamos nos referindo à **<font color=darkgreen> possibilidade de ocorrência de um dado evento**.

Imagine, por exemplo, um dado. Se quisermos calcular a probabilidade de o número 1 sair em uma jogada, devemos saber quais são todos os eventos possíveis, e, dentro desta situação, medir a chance de o número 1 aparecer. Neste caso, é uma em seis.

Se quisermos saber a probabilidade de o número 1 <font color = #21618C>**ou**<font color = black> o número 2 saírem, devemos **<font color = #21618C>somar**<font color = black> as probabilidades. O mesmo processo funciona para qualquer soma com outros números.

### <font color = #21618C> Uma Distribuição de Probabilidade ocorre quando  :

<br>
    
1º : 0 >= Probabilidade do evento <= 1
<br>
               
2º : Quando a soma de todos os eventos possíveis for igual a 1

# <font color = #21618C> Distribuição Binomial


<br>




A Distribuição Binomial se dá quando temos um evento em que existem somente duas possibilidades : <font color = #21618C> **sucesso<font color=black> ou <font color = #21618C>fracasso**.

Para ser uma Distribuição Binomial,**<font color = #21618C>os eventos devem ser independentes** <font color=black>uns dos outros ( o resultado de um evento não afeta o outro) 

E **<font color = #21618C>a probabilidade de sucesso permanece a mesma**  <font color=black>para todos os eventos

Um possível exemplo é o caso de lançar uma moeda, só haverão dois resultados possíveis, cara ou coroa. 


<br>




## <font color=green> Exercício I 
    
<br>
    
Suponha que um restaurante que possui um Drive-Thru deseja  **<font color=green>medir a quantidade de pedidos que são entregues corretamente** <font color=black>aos seus clientes. A **<font color=green>taxa de entregas feitas corretamente é de 80%** <font color=black>. Em um dia em que três clientes comparecem ao Drive-Thru, **<font color=green>qual é a probabilidade de dois pedidos serem entregues corretamente ?**


<br>




A probabilidade de um pedido ser entregue incorretamente será 1 - 0.8 = <font color=purple>**0.2**

E possuímos <font color=purple>**3 situações** :

1 - O primeiro pedido é entregue incorretamente <font color=purple>**(icc)**

2 - O segundo pedido é entregue incorretamente <font color=purple> **(cic)**

3 - O terceiro pedido é entregue incorretamente <font color=purple>**(cci)**

A probabilidade do primeiro  pedido ser entregue incorretamente é : 0.2 * 0.8* 0.8 = 0.128


<br>




 Porém, **<font color=red>é possível também que o segundo ou o terceiro cliente recebam o pedido errado** <font color=black>. Então devemos fazer o cálculo para as três situações possíveis e efetuar a soma.

<br>

P (icc) = <font color=purple>**0.2** <font color=black> * 0.8* 0.8 = 0.128
    
P (cic) = 0.8 * <font color=purple>**0.2** <font color=black>* 0.8 = 0.128
    
P (cci) = 0.8 * 0.8* <font color=purple>**0.2** <font color=black> = 0.128

P de um pedido com erro (ou dois pedidos corretos) : **<font color=purple>icc + cic + cci = 38.4%**


<br>




Mas e se supormos uma situação em que são feitos **<font color=green>20 pedidos** <font color=black> e desejamos saber a **<font color=green>probabilidade de ocorrer a entrega de 18 pedidos corretamente, considerando uma probabilidade de 80% ?**

### <font color = green> Solução "Manual"


<br>




```python
# Importando uma função que calcula o fatorial

from math import factorial
```

```python
# Declarando os valores para o cálculo do exercício

x = 18
n = 20

permutacao = factorial(n)/(factorial(x)*factorial(n - x))
probab_s = 0.8
probab_f = 0.2 

# Calculando a probabilidade do evento ocorrer

probab_evento = (permutacao* probab_s**x)*(probab_f**(n-x))
probab_evento
```


<br>

# <font color = #21618C> Distribuição de Poisson

### <font color = #21618C> É considerada uma distribuição de Poisson quando : 

1. A probabilidade de uma ocorrência é a mesma em todo o intervalo observado.
<br>

2. O número de ocorrências em determinado intervalo é independente<font color = black> do número de ocorrências em outros intervalos.
<br>
3. A probabilidade de uma ocorrência é a mesma em intervalos de igual comprimento.

<img src = "Imagens/Fórmulas de Poisson .png" width = "800" align = "center" />

Onde:

$e$ = constante cujo valor aproximado é <font color = #21618C> 2,7182

<font color = black> $\mu$ = representa o <font color = #21618C> número médio de ocorrências <font color = black> em um determinado intervalo de tempo ou espaço

$k$ = <font color = #21618C> número de sucessos no intervalo<font color = black>  desejado

## <font color = green> Exercício III

<br>
    
Suponha que a Data.Aqa possui um cinema. <font color = black> Sabendo que **<font color = green>a média de ingressos comprados por sessão é de 135** <font color = black>, qual é a probabilidade de que, em um determinado dia, 170 pessoas queiram comprar um ingresso ? E qual é a probabilidade de mais do que 150 pessoas<font color = black> quererem comprar um ingresso ?

<br>


### <font color = green> Solução "Manual"

<br>





```python
# Importando biblioteca para realizar o cálculo

from decimal import Decimal
```

```python
# Declarando variáveis

euler = 2.7182
media = 135
k = 170
```

```python
# Realizando os cálculos do numerador e denominador da equação

numerador = Decimal(euler**(-media)) * Decimal((media**k))
denominador = factorial(170)

# Calculando a probabilidade do evento 

probab_poisson = numerador/denominador
probab_poisson
```

<br>





### <font color = green> Solução Automática


<br>




```python
# Importando a função Poisson

from scipy.stats import poisson 
```

```python
# Valores para o cálculo

k = 170
media = 135

# Probabilidade de 170 pessoas quererem comprar o ingresso

poisson.pmf(k, media)
```

O evento acima é pouco provável de acontecer

<br>

### <font color = green> Visualizando Graficamente 

<br>

```python
# Criando o eixo x 

x = np.arange(0, 200, 0.1)
 
# Criando, no eixo y, uma probabilidade para cada um dos valores do eixo x 

y = poisson.pmf(x, mu=135)

# Gerando o Gráfico
plt.plot(x, y)
 
# Configurando as legendas da imagem

plt.ylabel('Probabilidade', fontsize = 13)
plt.title('Função de Massa de Probabilidade Poisson ', fontsize=22)

# Configurando a formatação da imagem

fig = plt.gcf()
fig.set_size_inches(15, 8)

# Mostrando o gráfico 
plt.show()
```

```python
# Probabilidade de mais do que 150 pessoas quererem comprar o ingresso 

1 - poisson.cdf(150, media)
```

<font color = green>**Resposta** <font color = black> : A probabilidade de 170 pessoas comprarem ingressos em um determinado dia é 0.046366%. A probabilidadde de mais de 150 pessoas quererem comprar ingressos em um determinado dia é 9.27 %

# <font color = #21618C> Distribuição Normal

<br>

**<font color = #21618C> A Distribuição Normal** <font color = black>, também conhecida como Distribuição de Gauss, **<font color = #21618C>é a distribuição mais conhecida dentro da estatística** <font color = black>. Isso ocorre pois uma série de fenômenos (altura, QI,peso entre outros) se comportam como uma Distribuição Normal. 

<br>

### <font color = #21618C> Alguns Atributos da Distribuição Normal

<br>

1 - **<font color = #21618C>É simétrica** <font color = black>em torno da média

```python
# Importando a biblioteca seaorn para construir o gráfico da normal

import seaborn as sns
```

```python
# Função random.normal( Média, Desvio Padrão, Tamanho da Amostra )

x = np.random.normal(0, 1, 500000)
x = np.sort(x)
```

```python
# Visualizando a Distribuição Normal

color = ['#144775']
fig=plt.figure(figsize=(16,6))


sns.kdeplot(data=x,fill=True, common_norm=False, 
            palette="crest",alpha=.9, color =color,
            shade=True, linewidth=1, edgecolor='black', multiple='layer')

plt.title("Distribuição Normal",family='Serif', weight='bold', size=20)
plt.xlabel('Desvio Padrão', fontsize=(13))
plt.ylabel('Densidade', fontsize=(13))
plt.show()
```

# <font color = #21618C> O Teste Z 

<br>

O Teste Z é um teste estatístico que **<font color = #21618C>serve para verificar probabilidades e níveis de significância** <font color = black> de determinados eventos.

<img src = "Imagens/Teste Z.png" width = "600" />

Onde:

$x$ = variável observada 

$\sigma$ = desvio padrão

$\mu$ = média

<br>

### <font color = #21618C> O desvio padrão possui a mesma unidade dos dados 

```python
import pandas as pd
from scipy.stats import norm

tabela_normal_padronizada = pd.DataFrame(
    [], 
    index=["{0:0.2f}".format(i / 100) for i in range(0, 400, 10)],
    columns = ["{0:0.2f}".format(i / 100) for i in range(0, 10)])

for index in tabela_normal_padronizada.index:
    for column in tabela_normal_padronizada.columns:
        Z = np.round(float(index) + float(column), 2)
        tabela_normal_padronizada.loc[index, column] = "{0:0.4f}".format(norm.cdf(Z))

tabela_normal_padronizada.rename_axis('Z', axis = 'columns', inplace = True)

tabela_normal_padronizada
```
