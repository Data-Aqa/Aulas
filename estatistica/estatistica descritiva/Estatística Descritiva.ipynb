De maneira geral, podemos dividir o campo da estatística em 3 áreas: a **estatística descritiva**, a **probabilidade** e a **estatística inferencial**.

# <font color = #21618C> <u>Tipos de Dados</u>

Assim como quando programando em python é importante que saibamos o tipo de dado com o qual estamos lidando, na estatística também devemos ter esse cuidado. Isso porque o tipo de dado irá determinar como vamos lidar com ele e quais técnicas devemos utilizar.

##  <font color = #21618C><u>Numéricos</u>

# <font color = #21618C> <u>Estatística Descritiva</u>
---
Agora que entendemos quais são os tipos de variáveis possíveis, vamos entrar de fato na estatística.

No começo do notebook, dissemos que a estatística descritiva foca em descrever e sumarizar um conjunto de dados. Ela faz isso por meio de uma série de medidas que trazem informações sobre a disposição dos dados.

Essas medidas se dividem em medidas de <font color = #228c57>__Tendência Central__<font color = black>, de <font color = #228c57>__Dispersão__ e __Separatrizes__.

## <font color = #21618C> <u>Medidas de Tendência Central</u>

As medidas de tendência central são medidas que, de certa forma, descrevem o conjunto de dados como um todo. Elas capturam os valores centrais de um conjunto de dados.

São a <font color = #228c57>__média__<font color = black>,<font color = #228c57> __mediana__<font color = black>, e <font color = #228c57>__moda__.

### <font color = #21618C><u>Média</u>

A média é a medida que descreve a _concentração_ dos dados dentro de um conjunto ou distribuição, e é obtida pela fórmula abaixo

Continuando com o conjunto de dados _HR Analytics_, vamos agrupar os dados por __departamento__ e calcular a média das colunas.

```python
lista = dados.columns
```

```python
lista
```

```python
dados.groupby('department')[lista].mean()  # .mean(): calcula a média de todos os valores no eixo (axis) desejado
```

### <font color = #21618C><u>Mediana</u>

A mediana é a medida que nos mostra o valor _central_ do conjunto de dados. Obtemos ela organizando os dados em ordem _crescente_ e pegando o elemento que divide o conjunto no meio.

Ela é muito usada por ser menos sensível à valores extremos do que a média.

Vamos comparar a média e a mediana da coluna `average_monthly_hours` por departamento para ver se algum deles tem valores extremos.

```python
dados.groupby('department')[['average_monthly_hours']].agg([np.mean, np.median])

#.agg(): função que permite fazer a agregação da uma linha ou coluna utiliizando uma ou mais operações.
```

Podemos ver que em todos os departamentos a média e mediana não diferem muito.

Para exemplificar como valores extremos podem afetar a média e mediana, vamos imaginar um país com apenas dez habitantes, cujos salários são dados pela lista abaixo:

```python
salarios = [1000, 1000, 1000, 2000, 6000, 4500, 3500, 1200, 5000, 100000]
```

```python
media_salarial = np.mean(salarios)
print(media_salarial)
```

Olhando para os salários e para a média salarial, a média é uma boa representação dos salários do país? Não!

Isso ocorre porque temos um valor extremo no conjunto de dados que acaba puxando o valor da média para cima. Para termos uma representação mais fiel dos salários, usamos a mediana.

```python
mediana_salarial = np.median(salarios)
print(mediana_salarial)
```

Como podemos ver, nesse caso a mediana representa melhor o conjunto de dados do que a média, mas isso não significa que a média esteja errada! Por isso é importante entender o que os dados representam para saber o melhor jeito de descreve-los.

### <font color = #21618C><u>Moda</u>

A moda é a medida que nos mostra o valor mais frequente do conjunto de dados. Ela é obtida contando quantas vezes cada elemento se repete no conjunto de dados e selecionando aquele que aparece com maior frequência.

Vamos calcular a moda da coluna `department` para vermos qual é o departamento com mais funcionários.

```python
def moda(coluna):
    departamento = dados[coluna].value_counts().idxmax()
    contagem = dados[coluna].value_counts().max()
    return print("O departamento mais frequente é {}, com {} funcionários.".format(departamento, contagem))

# .value_counts(): retorna uma série com cada categoria (índice) e quantas vezes ela aparece (valores)
# .value_counts().idmax(): retorna o índice do valor que mais aparece
# .value_counts().max(): retorna o valor que mais se repete
```

```python
moda('department')
```

Um uso comum da moda é preencher valores categórios faltantes do conjunto de dados.

## <font color = #21618C><u>Medidas de Dispersão</u>
---
As medidas de dispersão complementam as medidas de tendência central. Elas nos informam sobre como os dados se dispersam ao redor da média.

As duas medidas de dispersão são a __variância__ e o __desvio padrão__.

### <font color = #21618C><u>Variância</u>

A variância é difinida como _a distância média_ de cada ponto do conjunto de dados até a média do conjunto. Sua fórmula é:

<center>$VAR = \sigma^2 = \frac{\sum_{i=1}^{n}(x_i - \mu)^2} {n}$</center>

Nós interpretamos a variância como quão distante da média está cada ponto do conjunto de dados. Quanto maior a variância de um conjunto de dados, mais espalhados estão os pontos.

### <font color = #21618C><u>Desvio Padrão</u>

O desvio padrão, assim como a variância, mede quão afastados da média estão cada ponto do conjunto de dados. Ele é definido como a raiz quadrada da média da distância de cada ponto do conjunto de dados até a média do conjunto. Sua fórmula é:

<center>$\sigma = \sqrt\frac{\sum_{i=1}^{n}(x_i - \mu)^2} {n}$</center>

Como podemos ver, o __desvio padrão__ é a raiz quadrada da __variância__. A diferença entre as duas medidas é que a variância, por ser elevada ao quadrado, é mais sensível a valores muito distantes da média. Além disso, é mais fácil interpretar o desvio padrão, por ele ter a mesma unidade da média enquanto a variância tem a unidade elevada ao quadrado.

## <font color = #21618C><u>Medidas Separatrizes</u>
---
Medidas separatrizes são medidas usadas para separar o conjunto de dados em partes. A mediana, que vimos nas medidas de tendência central, é uma medida separatriz: ela divide o conjunto de dados no meio.

As medidas separatrizes mais comuns são os _quartis_. São as medidas que dividem o conjunto de dados em quatro partes iguais.

O primeiro quartil, ou quartil inferior, é o valor aos 25% da amostra ordenada. O segundo quartil, mediana, está aos 50% e o terceiro quartil - quartil superior - aos 75%.

Podemos ver todas as medidas apresentadas até o momento utilizando o método `.describe()`

# <font color = #21618C><u>Visualizações</u>

Uma grande aliada da estatística descritiva são as visualizações. Por meio delas, conseguimos ver graficamente como os dados estão dispersos no conjunto de dados.

Vamos falar e exemplificar 3 tipos de gráficos usados para visualização: os **Boxplots**, **Histogramas** e **Bar Plots** (ou gráficos de barra).

### <font color = #21618C><u>Boxplot</u>
---
Os Boxplots são utilizados para ver a distribuição dos dados por meio de quartis, e é muito útil para detectar _outliers_.

```python
fig, ax = plt.subplots()

ax.boxplot(dados['average_monthly_hours'])

ax.axhline(np.percentile(dados['average_monthly_hours'], 75), 
           linestyle = '--',  linewidth = 1,
           c = 'b', label = '3º Quartil')

#ax.axhline(np.median(dados['average_monthly_hours']),
#           linestyle = '--', linewidth = 1,
#           c = 'k', label = '2º Quartil (mediana)')

ax.axhline(np.percentile(dados['average_monthly_hours'], 25),
           linestyle = '--', linewidth = 1,
           c = 'r', label = '1º Quartil')

ax.annotate('valor mínimo', xy = (.95, dados['average_monthly_hours'].min()),
            xytext = (0.6, dados['average_monthly_hours'].min()),
            arrowprops = dict(facecolor = 'black', width = 2, headwidth = 8))

ax.annotate('valor máximo', xy = (1.01, 304),
            xytext = (1.1, 266),
            arrowprops = dict(facecolor = 'black', width = 2, headwidth = 8))

ax.annotate('mediana', xy = (1.1, dados['average_monthly_hours'].median()),
            xytext = (1.2, 199),
            arrowprops = dict(facecolor = 'black', width = 2, headwidth = 8))

ax.legend()

plt.show()


# ax.axhline(): cria uma linha horizontal no gráfico no ponto passado como argumento
# np.percentile(dados, percentil_desejado): retorna o valor correspondente ao percentil passado como argumento
# ax.legend(): exibe as legendas definidas pelos argumentos 'label'
# ax.annotate(): faz anotações nas coordenadas especificadas nos argumentos
```

Pelo boxplot podemos ver que 50% dos dados (espaço entre o 3º e 1º quartis) estão entre 150 e 250 horas mensais.

### <font color = #21618C><u>Histogramas</u>
---
O histograma é um tipo de gráfico utilizado para representar a distribuição de frequências de um conjunto de dados, e pode ser utilizado para representar variáveis númericas, tanto contínuas quanto discretas.

Ele faz isso dividindo o intervalo da variável escolhida em  "bins" igualmente espaçadas. Depois disso conta quantas observações caem dentro de cada uma das `bins`.

Para exemplificar, vamos olhar a distribuição de frequências das horas mensais trabalhadas.

```python
fig, ax = plt.subplots(figsize = (12, 5))

ax.hist(dados['average_monthly_hours'], ec = 'black')
ax.grid(True, alpha = .25) 
ax.set_title('Horas mensais trabalhadas', fontsize = 15)
ax.set_xlabel('Horas trabalhadas') 
ax.set_ylabel('Frequência') 

plt.show()

#ax.grid(): permite controlar parâmetros das grades do gráfico
#ax.set_title(): define o título do gráfico
#ax.set_xlabel: define o nome do eixo x
#ax.set_tlabel: defina o nome do eixo y
```

### <font color = #21618C><u>Barplots</u>
---
Barplots, os famosos gráficos de barra também servem para visualizarmos a distribuição de frequência de uma variável, porém utilizamos ele para variáveis categóricas.

Vamos visualizar a distribuição de frequências dos departamentos

```python
fig, ax = plt.subplots(figsize = (12, 6))

ax.bar(dados['department'].unique(), dados['department'].value_counts(), color = '#21618C')
ax.grid(True, axis = 'y', alpha = .25)
ax.set_ylabel('Quantidade')
ax.set_title('Quantidade de funcionários por departamento', fontsize = 15)

plt.show()

# ax.bar():
```

## <font color = #21618C><u>Plot extra!</u>
---
Quando falamos sobre histogramas anteriormente, mostramos a distribuição das horas mensais trabalhadas de toda a empresa. Mas e se nós quisessemos ver esta mesma distribuição para **cada** depertamento?

```python
departments = dados.set_index('department')[['average_monthly_hours']]
```

```python
fig, axes = plt.subplots(4, 3, figsize = (15, 9))
axs = axes.flat 
for k in [10, 11]:
    fig.delaxes(axs[k])

departamentos = departments.index.unique()

for i in range(0, (len(axs)-2)):
    axs[i].hist(departments.loc[departamentos[i]], ec = 'black')
    
    axs[i].axvline(departments.loc[departamentos[i]]['average_monthly_hours'].mean(),
                  color = 'red', linestyle = '--', linewidth = 1, label = 'média')
    
    axs[i].axvline(departments.loc[departamentos[i]]['average_monthly_hours'].median(),
                  color = 'k', linestyle = '--', linewidth = 1, label = 'mediana')
    
    axs[i].set_title("{}".format(departments.index.unique()[i]))
    
    

fig.text(.5, .1, 'Horas Trabalhadas', ha = 'center', va = 'center', fontsize = 12)
fig.text(.04, .5, 'Quantidade de funcionários', ha = 'center', va = 'center', rotation = 'vertical', fontsize = 12)
fig.suptitle('Horas mensais trabalhadas por departamento', fontsize = 15)
fig.subplots_adjust(left = .1, hspace = .5)

plt.show()

# plt.subplots(nº lin, nº col): retorna um objeto .figure com (nºlin * nºcol) eixos
# .flat: "achata" o vetor em que é chamado
# ax.axvline(): cria uma linha vertical no ponto passado como argumento
# fig.text(): permite colocar textos na figura
# fig.suptitle(): define o título da figura
# fig.subplots_adjust(): define o ajustamento dos eixos (axes) dentro da figura
```
